```python
#13
import matplotlib.pyplot as plt
import seaborn as sns

correlation_matrix = data.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

Тепловая карта  парных зависимостей признаков.

Видим, что зависимость наблюдается между признаками

- citric acid ~ fixed acidity
- density ~ fixed acidity
- pH ~ fized acidity
- total sulfur dioxide ~ free sulfur dioxide

Однако коэффициент корелляции не слишком высокий, все признаки можем оставить

```python
#14
from sklearn.decomposition import PCA

pca = PCA()
pca.fit(X_train, y_train)

explained_variance_ratio_cumsum = pca.explained_variance_ratio_.cumsum()
n_components = (explained_variance_ratio_cumsum < 0.90).sum() + 1
n_components
```

для объяснения 90% дисперсии будет достаточно 7 компонент

```python
#15
max_contributor_feature = data.columns[pca.components_[0].argmax()]
max_contributor_feature
```

Признак, который вносит наибольший вклад в первую компоненту -- fixed acidity

```python
plt.figure(figsize=(10, 6))
plt.plot(explained_variance_ratio_cumsum, marker='o', linestyle='-')
plt.title('Explained Variance Ratio Cumulative Sum')
plt.xlabel('Number of Components')
plt.ylabel('Explained Variance Ratio Cumulative Sum')
plt.grid(True)
plt.show()

# Визуализация PCA компонент
plt.figure(figsize=(10, 6))
plt.bar(range(len(pca.components_[0])), pca.components_[0])
plt.title('PCA Components')
plt.xlabel('Feature Index')
plt.ylabel('PCA Component')
plt.xticks(range(len(pca.components_[0])), data.columns, rotation=90)
plt.grid(True)
plt.show()
```

### Статистические оценки параметров и гистограммы распределений

```python
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC

dataset = pd.concat([data, y], axis = 1)

# Статистика и графики
print(dataset.describe())

# Гистограммы
dataset.hist(figsize=(12,10))
plt.tight_layout()
plt.show()
```

### Построение регрессии для предсказания качества

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Обучение модели линейной регрессии
linear_reg = LinearRegression()
linear_reg.fit(X_train, y_train)

# Предсказание на тестовой выборке
y_pred_linear = linear_reg.predict(X_test)

# Оценка качества модели
mse = mean_squared_error(y_test, y_pred_linear)
r2 = r2_score(y_test, y_pred_linear)
print("Mean Squared Error:", mse)
print("R^2 Score:", r2)
```

Качество линейной регрессии низкое. Посмотрим на scatter plot для каждого признака по всем объектам выборки. Это позволит нам визуально оценить взаимосвязь между признаками и целевой переменной  

```python
plt.figure(figsize=(20, 12))
for i, feature in enumerate(data.columns):
    plt.subplot(3, 4, i + 1)
    plt.scatter(dataset[feature], dataset['quality'], alpha=0.5)
    plt.xlabel(feature)
    plt.ylabel('Quality')
    plt.title(f'Scatter plot for {feature}')
plt.tight_layout()
plt.show()
```

Оценив графики, сделаем вывод, что модель линейной регрессии будет слишком простой для данной задачи. Построим random forest regression

```python
from sklearn.ensemble import RandomForestRegressor

rf_regressor = RandomForestRegressor(n_estimators=600, random_state=42)

rf_regressor.fit(X_train, y_train)
y_pred_rf = rf_regressor.predict(X_test)

mse_rf = mean_squared_error(y_test, y_pred_rf)
r2_rf = r2_score(y_test, y_pred_rf)

print("Random Forest Mean Squared Error:", mse_rf)
print("Random Forest R^2 Score:", r2_rf)
```

видим, качество модели невыское, подберем гиперпараметры


```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
rf_regressor = RandomForestRegressor(random_state=42)
grid_search = GridSearchCV(estimator=rf_regressor, param_grid=param_grid, cv=5, scoring='neg_mean_squared_error', n_jobs=-1)
grid_search.fit(X_train, y_train)

print("Best Parameters:", grid_search.best_params_)
print("Best Negative Mean Squared Error:", grid_search.best_score_)
```

```python
rf_reg_best = RandomForestRegressor(max_depth=None, min_samples_leaf = 1, min_samples_split=2, n_estimators = 200, random_state=42)

rf_reg_best.fit(X_train, y_train)
rf_reg_pred = rf_reg_best.predict(X_test)

mse_rf_best = mean_squared_error(y_test, rf_reg_pred)
r2_rf_best = r2_score(y_test, rf_reg_pred)

print("Random Forest Mean Squared Error:", mse_rf_best)
print("Random Forest R^2 Score:", r2_rf_best)
```

Видим, что модели регрессии работают плохо, перейдем к методам классификации.

### Построение классификации для предсказания качества

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.neighbors import KNeighborsClassifier
# K ближайших соседей (KNN)
knn_classifier = KNeighborsClassifier()
knn_classifier.fit(X_train, y_train)
y_pred_knn = knn_classifier.predict(X_test)

accuracy_knn = accuracy_score(y_test, y_pred_knn)
precision_knn = precision_score(y_test, y_pred_knn, average='weighted')
recall_knn = recall_score(y_test, y_pred_knn, average='weighted')
f1_knn = f1_score(y_test, y_pred_knn, average='weighted')

print("KNN Metrics:")
print("Accuracy:", accuracy_knn)
print("Precision:", precision_knn)
print("Recall:", recall_knn)
print("F1 Score:", f1_knn)
```

```python
# логистическая регрессия
from sklearn.linear_model import LogisticRegression

log_reg_classifier = LogisticRegression(max_iter=1000)
log_reg_classifier.fit(X_train, y_train)
y_pred_log_reg = log_reg_classifier.predict(X_test)

accuracy_log_reg = accuracy_score(y_test, y_pred_log_reg)
precision_log_reg = precision_score(y_test, y_pred_log_reg, average='weighted')
recall_log_reg = recall_score(y_test, y_pred_log_reg, average='weighted')
f1_log_reg = f1_score(y_test, y_pred_log_reg, average='weighted')

print("\nLogistic Regression Metrics:")
print("Accuracy:", accuracy_log_reg)
print("Precision:", precision_log_reg)
print("Recall:", recall_log_reg)
print("F1 Score:", f1_log_reg)
```

```python
#случайный лес

from sklearn.ensemble import RandomForestClassifier

rf_classifier = RandomForestClassifier(n_estimators=300, random_state=42)
rf_classifier.fit(X_train, y_train)

y_pred_rf = rf_classifier.predict(X_test)

accuracy_rf = accuracy_score(y_test, y_pred_rf)
precision_rf = precision_score(y_test, y_pred_rf, average='weighted')
recall_rf = recall_score(y_test, y_pred_rf, average='weighted')
f1_rf = f1_score(y_test, y_pred_rf, average='weighted')

print("\nRandom Forest Metrics:")
print("Accuracy:", accuracy_rf)
print("Precision:", precision_rf)
print("Recall:", recall_rf)
print("F1 Score:", f1_rf)
```

Видим, что показатели метрик у классификатора RandomForest лучше, чем у остальных, однако их значения не высоки, подберем гиперпараметры.

```python
param_grid = {
    'n_estimators': [100, 200, 250, 300],
    'max_depth': [5, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

rf_classifier = RandomForestClassifier(random_state=42)

grid_search = GridSearchCV(estimator=rf_classifier, param_grid=param_grid, cv=3, n_jobs=-1)
grid_search.fit(X_train, y_train)

print("Best Parameters:", grid_search.best_params_)
print("Best Score:", grid_search.best_score_)
```

Лучший результат по acuracy - 0.66 у исходного RF классификатора.

### Бинарный классификатор

```python
def binarize_labels(y):
    return np.where(y > 5, 1, 0)
    
y_train_binary = binarize_labels(y_train)
y_test_binary = binarize_labels(y_test)

rf_binary_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
rf_binary_classifier.fit(X_train, y_train_binary)
y_pred_binary = rf_binary_classifier.predict(X_test)

binary_accuracy = accuracy_score(y_test_binary, y_pred_binary)
binary_precision = precision_score(y_test_binary, y_pred_binary)
binary_recall_ = recall_score(y_test_binary, y_pred_binary)
binary_f1 = f1_score(y_test_binary, y_pred_binary)
print("Binary Classifier Accuracy:", binary_accuracy)
print("Binary Classifier Precision", binary_precision)
print("Binary Classifier Recall:", binary_recall_)
print("Binary Classifier F1 Score:", binary_f1)
```

```python
data3 = dataset.drop([199, 353, 900, 1124, 1239, 1423, 1484])
data3.shape
```

```python
data3.head()
```

```python
from sklearn.model_selection import train_test_split

y = data3['quality']
data3 = data3.drop('quality', axis = 1)

X_train, y_train, X_test, y_test = train_test_split(data3, y, test_size = 0.3, random_state = 42)
```

```python
from sklearn.linear_model import Lasso, Ridge
from sklearn.metrics import mean_squared_error, r2_score

lasso_reg = Lasso(alpha=0.001) 
lasso_reg.fit(X_train, y_train)
y_pred_lasso = lasso_reg.predict(X_test)

mse_lasso = mean_squared_error(y_test, y_pred_lasso)
r2_lasso = r2_score(y_test, y_pred_lasso)

print("MSE:", mse_lasso)
print("R2:", r2_lasso)
```

```python
ridge_reg = Ridge(alpha = 0.5)
ridge_reg.fit(X_train, y_train)
y_pred_ridge = ridge_reg.predict(X_test)

mse_ridge = mean_squared_error(y_test, y_pred_ridge)
r2_ridge = r2_score(y_test, y_pred_ridge)

print("MSE:", mse_ridge)
print("R2:", r2_ridge)
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LinearRegression

poly = PolynomialFeatures(degree=2)
poly_reg = make_pipeline(poly, LinearRegression())
poly_reg.fit(X_train, y_train)
y_pred_poly = poly_reg.predict(X_test)

# Оценка качества модели
mse_poly = mean_squared_error(y_test, y_pred_poly)
r2_poly = r2_score(y_test, y_pred_poly)
print("MSE:", mse_poly)
print("R2:", r2_poly)
```

степени выше 2 работают некорректно, качество модели не улучшилось

### Градиентный бустинг

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error

gradient_boosting = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, random_state=42)
gradient_boosting.fit(X_train, y_train)

y_pred_gb = gradient_boosting.predict(X_test)
mse_gb = mean_squared_error(y_test, y_pred_gb)
r2_gb = r2_score(y_test, y_pred_gb)

print("MSE:", mse_gb)
print("r2:", r2_gb)
```

**Текущие гиперпараметры для модели подбирались отдельно и выдают лучшие результаты**

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score

gradient_boosting = GradientBoostingClassifier(n_estimators=300, learning_rate=0.25, random_state=42)
gradient_boosting.fit(X_train, y_train)

y_pred_gb = gradient_boosting.predict(X_test)
accuracy_gb = accuracy_score(y_test, y_pred_gb)

print("Gradient Boosting Accuracy:", accuracy_gb)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=7)
X_pca = pca.fit_transform(data3)

X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca, y, test_size=0.3, random_state=42)

gradient_boosting = GradientBoostingClassifier(n_estimators=350, learning_rate=0.25, random_state=42)
gradient_boosting.fit(X_train_pca, y_train)

y_pred_gb = gradient_boosting.predict(X_test_pca)

accuracy_gb = accuracy_score(y_test, y_pred_gb)
print("Gradient Boosting Accuracy with PCA:", accuracy_gb)
```
