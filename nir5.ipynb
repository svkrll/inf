```python
#11
y = data['quality']
data = data.drop(columns = ['quality'])
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()

data = pd.DataFrame(ss.fit_transform(data), columns = data.columns)
```

```python
data.describe()
```

```python
#10 - residual sugar
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data, y, test_size=0.3, random_state=42)
```

```python
#12
X_train.shape
```

```python
#13
import matplotlib.pyplot as plt
import seaborn as sns

correlation_matrix = data.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

Тепловая карта  парных зависимостей признаков.

Видим, что зависимость наблюдается между признаками

- citric acid ~ fixed acidity
- density ~ fixed acidity
- pH ~ fized acidity
- total sulfur dioxide ~ free sulfur dioxide

Однако коэффициент корелляции не слишком высокий, все признаки можем оставить

```python
#14
from sklearn.decomposition import PCA

pca = PCA()
pca.fit(X_train, y_train)

explained_variance_ratio_cumsum = pca.explained_variance_ratio_.cumsum()
n_components = (explained_variance_ratio_cumsum < 0.90).sum() + 1
n_components
```

для объяснения 90% дисперсии будет достаточно 7 компонент

```python
#15
max_contributor_feature = data.columns[pca.components_[0].argmax()]
max_contributor_feature
```

Признак, который вносит наибольший вклад в первую компоненту -- fixed acidity

```python
plt.figure(figsize=(10, 6))
plt.plot(explained_variance_ratio_cumsum, marker='o', linestyle='-')
plt.title('Explained Variance Ratio Cumulative Sum')
plt.xlabel('Number of Components')
plt.ylabel('Explained Variance Ratio Cumulative Sum')
plt.grid(True)
plt.show()

# Визуализация PCA компонент
plt.figure(figsize=(10, 6))
plt.bar(range(len(pca.components_[0])), pca.components_[0])
plt.title('PCA Components')
plt.xlabel('Feature Index')
plt.ylabel('PCA Component')
plt.xticks(range(len(pca.components_[0])), data.columns, rotation=90)
plt.grid(True)
plt.show()
```

```python
from sklearn.manifold import TSNE

tsne = TSNE(n_components=2, random_state=42)
data_tsne = tsne.fit_transform(X_train, y_train)

plt.figure(figsize=(8, 6))
plt.scatter(data_tsne[:, 0], data_tsne[:, 1], alpha=0.5)
plt.title('t-SNE Visualization')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.show()
```

Кластеры представляют собой группы точек, которые находятся близко друг к другу в двухмерном пространстве. Визуально выборка не разделяется на кластеры, т.к. нет сильно обособленных групп точек.
