```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("BankChurners.csv")
data.head(10)
```

## Предобработка данных

```python
data.isna().sum()
#пропусков в данных нет
```

```python
y = data['Total_Trans_Amt']
data = data.drop(columns=['Total_Trans_Amt'])
y.head(), data.columns
```

разделим признаки на категориальные и числовые

```python
data.dtypes
```

```python
cat_feature_mask = (data.dtypes == 'object')
X_cat = data[data.columns[cat_feature_mask]]
X_num = data[data.columns[~cat_feature_mask]]
data.shape, X_cat.shape, X_num.shape
```

Закодируем категориальные переменные. Для признаков **Attrition_FLag, Gender, Marital_Status, Card_Category** будем использовать dummy-кодирование, для признака **Income_Category** -- label-кодирование

### from sklearn.preprocessing import LabelEncoder
for col in X_cat.columns:
    print(col, X_cat[col].unique())

```python
X_cat = pd.get_dummies(X_cat, columns=['Education_Level','Attrition_Flag','Gender','Marital_Status', 'Card_Category'], drop_first=True)
X_cat.head()
```

```python
X_cat['Income_Category'].unique()
```

```python
income_cats = ['Unknown', 'Less than $40K', '$40K - $60K', '$60K - $80K', '$80K - $120K', '$120K +']
dict = {x: i for i, x in enumerate(income_cats)}
X_cat['Income_Category'] = X_cat['Income_Category'].map(dict)
```

```python
X_cat = X_cat.astype(int)
X_cat.head(10)
```

Перейдем к обработке числовых признаков. Удалим столбец **CLIENTNUM**, так как он является идентификационным и нормализуем данные, используя **StandardScaler**

```python
X_num.head()
```

```python
X_num = X_num.drop(columns=['CLIENTNUM'])
X_num.head()
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()

X_num = pd.DataFrame(ss.fit_transform(X_num), columns = X_num.columns)
```

```python
X_num.describe()
```

Зададим классы на y. Объединим X_cat и X_num в одну выборку и разделим её на тестовую и на обучающую

```python
from sklearn.model_selection import train_test_split
X = pd.concat([X_cat, X_num], axis = 1)
y = y.apply(lambda x: 0 if x <= 5000 else 1)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=123)
```

## Построение классификатора

Построим логистическую регрессию

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, f1_score, recall_score

lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
y_test.value_counts() 
```

Классы несбалансированы

```python
accuracy_score(y_test, lr.predict(X_test))
```

```python
precision_score(y_test, lr.predict(X_test))
```

```python
recall_score(y_test, lr.predict(X_test))
```

```python
f1_score(y_test, lr.predict(X_test))
```

Accuracy, доля правильных ответов - 0.93

Precision, точность предсказания - 0.89, т.е. в уровень доверия классификатору при a(x) = 1 89% 

Recall, полнота модели - 0.70, т.е. модель находит 70% объектов положительного класса

f1_score, гармоническое среднее полноты и точности - 0.79. так как выборка несбалансированная и отрицательного класса гораздо больше, чем положительного, значение F-меры при &beta; = 1, показывает насколько хороша модель, насколько точно модель показывает результаты, которые имеют для нас значение

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators = 100, criterion = 'entropy')
rf.fit(X_train, y_train)
```

```python
accuracy_score(y_test, rf.predict(X_test))
```

```python
precision_score(y_test, rf.predict(X_test))
```

```python
recall_score(y_test, rf.predict(X_test))
```

```python
f1_score(y_test, rf.predict(X_test))
```

```python
from sklearn.model_selection import GridSearchCV
rf_grid = RandomForestClassifier()

param_grid = {
    'n_estimators': [100, 150, 200, 250, 300, 350, 400, 450, 500]}

grid_search = GridSearchCV(estimator=rf_grid, param_grid=param_grid)
grid_search.fit(X_train, y_train)

#признаков много, большой GridSearch будет считаться очень долго, поэтому переберем только число деревьев
```

```python
grid_search.best_params_ 
```

```python
param_grid2 = {
    'n_estimators': [210, 220, 230, 240, 250, 260, 270, 280, 290]}

grid_search2 = GridSearchCV(estimator=RandomForestClassifier(), param_grid=param_grid2)
grid_search2.fit(X_train, y_train)
```

```python
grid_search2.best_params_ 
```

```python
rf_best = grid_search2.best_estimator_
```

```python
accuracy_score(y_test, rf_best.predict(X_test))
```
